```python
import pandas as pd
import numpy as np

data = pd.read_csv("C:/Users/manoj/OneDrive/Desktop/TASK/student_data.csv")
data = data.drop(['Dalc', 'famsize', 'Walc','health','famrel','freetime','G1','G2','G3','goout','romantic'], axis=1)


data.head(5)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))

plt.hist(data['age'], bins=5, alpha=0.5, label='Age', color='blue')
plt.hist(data['sex'], bins=5, alpha=0.5, label='sex', color='red')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Age and sex')
plt.legend()
plt.show()
```
